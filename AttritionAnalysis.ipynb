```python
df = df.drop(['EmployeeCount', 'Over18', 'StandardHours', 'PerformanceRating', 'EmployeeNumber'], axis=1)
```

```python
corr = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(corr, mask=mask, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, linewidths=.2, 
            fmt='.2f', annot_kws={"size": 8}, ax=ax)
```

```python
df_num = df.select_dtypes('int64')
```

```python
plt.rcParams.update({'figure.max_open_warning': 0})
for column in df_num:
    plt.figure()
    sns.histplot(x=df[column]) 
```
